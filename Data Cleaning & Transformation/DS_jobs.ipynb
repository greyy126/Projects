```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/greshashah126/Downloads/archive (8)/Uncleaned_DS_jobs.csv', low_memory=False )
pd.set_option("display.max_columns", None)
df
```

```python
#Checking for null values
df.isnull().sum()
```

```python
df['Salary Estimate'].dtype
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.rstrip("(Glassdoor est.)")
df['Salary Estimate'] = df['Salary Estimate'].str.strip("(Employ')")
```

```python
#Convert salary Estimates in int
def convert_salary(s):
    s = s.replace('$', '').replace('K', '').split('-')
    lower = int(s[0]) * 1000
    upper = int(s[1]) * 1000
    return (lower + upper) // 2  # Taking the average of the range

df['Salary Estimate'] = df['Salary Estimate'].apply(convert_salary)

print(df['Salary Estimate'].dtype)
```

```python
df.rename(columns={"Salary Estimate": "Salary Estimate (AVG)"}, inplace=True)
```

```python
df['Company Name'].unique()
```

```python
def remove_num(x):
    if "\n" in x:
        return x[:-3]
    else:
        return x
    
df['Company Name'] = df['Company Name'].apply(remove_num)
```

```python
df["Company Name"].unique()
```

```python
df['Company Name'] = df['Company Name'].str.strip('\n')
df
```

```python
#Removing not needed -1 values
def remove_negative_int_col(y):
    if y == -1: 
        return "NA"
    else:
        return y
    
df['Founded'] = df['Founded'].apply(remove_negative_int_col)


def remove_negative_string_col(z):
    if "-1" in z:
        return "NA"
    else:
        return z
    
df['Size'] = df['Size'].apply(remove_negative_string_col)
df['Headquarters'] = df['Headquarters'].apply(remove_negative_string_col)
df['Type of ownership'] = df['Type of ownership'].apply(remove_negative_string_col)
df['Industry'] = df['Industry'].apply(remove_negative_string_col)
df['Sector'] = df['Sector'].apply(remove_negative_string_col)
df['Revenue'] = df['Revenue'].apply(remove_negative_string_col)
df['Competitors'] = df['Competitors'].apply(remove_negative_string_col)

df
```

```python
#Sort Rating
df.Rating = np.where(df.Rating==-1.0,0,df.Rating)
df = df.sort_values(by='Rating', ascending=True)
```

```python
df['Size'] = df['Size'].str.replace('to', '-')
df['Size'] = df['Size'].str.replace('employees', '')
```

```python
#Splitting minimum and maximum employees
df['Size (Min Employees)']=df['Size'].apply(lambda x:x.split('-')[0])
df['Size (Max Employees)']=df['Size'].apply(lambda x:x.split('-')[-1])
df
```

```python
df['Size'] = df['Size'].replace("Unknown", "NA")
df['Type of ownership'] = df['Type of ownership'].replace("Unknown", "NA")
df['Revenue'] = df['Revenue'].replace("Unknown / Non-Applicable", "NA")
df['Size (Min Employees)'] = df['Size (Min Employees)'].replace("Unknown", "NA")
df['Size (Max Employees)'] = df['Size (Max Employees)'].replace("Unknown", "NA")
```
