```python
import pandas as pd
import numpy as np
data = pd.read_csv("/Users/greshashah126/Desktop/fifa21 raw data v2.csv", low_memory=False)
pd.set_option("display.max_columns", None)

data
```

```python
#Dropping Duplicates
data = data.drop_duplicates()
data = data.drop(columns = "Name")
data
```

```python
data.rename(columns={'LongName': 'Name'}, inplace=True)

data
```

```python
data = data.drop(columns = "photoUrl")
data = data.drop(columns = "playerUrl")
data.drop(columns="Loan Date End", inplace=True)
data
```

```python
#Checking for null values
data.isnull().sum()
```

```python
data = data.fillna("")
data
```

```python
#Stripping unwanted things
data['Club'] = data['Club'].str.lstrip("\n")

#Removing star
def remove_star(x):
    x = x.replace("★", "")
    return x
data['W/F'] = data['W/F'].apply(remove_star)
data['SM'] = data['SM'].apply(remove_star)
data['IR'] = data['IR'].apply(remove_star)
data
```

```python
data.Height.unique()
data['Height'].dtype
```

```python
#Converting height to cm
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"", "")
        feet = int(x[0]) * 30.48
        inch = int(x[2:]) * 2.54
        return round(feet+inch)
    elif x[-1] == "m":
        return int(x[:-2])

data['Height'] = data['Height'].apply(convert_height)
```

```python
#Converting weight to kg
def convert_weight(y):
    if 'lbs' in y:
        y = y.replace('lbs','')
        y = float(y) * 0.453592
        y = int(y)
        return y
    else:
        return int(y[:-2])

data['Weight'] = data['Weight'].apply(convert_weight)
```

```python
data['Weight'].dtype
```

```python
#Renaming height and weight columns
data.rename(columns={'Height': 'Height in cm'}, inplace=True)
data.rename(columns={'Weight': 'Weight in kg'}, inplace=True)

data
```

```python
data['Contract'].unique()
```

```python
#Simplifying contract column
def contract_status(a):
    if "On Loan" in a:
      a =  "ON LOAN"
      return a
    elif "~ 2020" in a:
      a = "CONTRACT OVER"
      return a
    elif "~ 2021" in a:
      a = "CONTRACT OVER"
      return a
    elif "~ 2022" in a:
      a = "CONTRACT OVER"
      return a
    elif "Free" in a:
      a = "NOT ON CONTRACT"
      return a
    else:
      return "ONGOING"
    
data['Contract'] = data['Contract'].apply(contract_status)

#Renaming contract column  
data.rename(columns={'Contract': 'Contract Status'}, inplace=True)

data
```

```python
data['Contract Status'].unique()
```

```python
#Splitting Joined Column
data[['Month', 'Date', 'Year']]= data['Joined'].str.split(" ", expand = True)
data = data.drop(columns = "Joined")
data['Date'] = data['Date'].str.rstrip(",")

data
```

```python
#Converting value and release clause
def convert_Col_Value_RC(x):
    if "€" in x:
        return x[1:]
    elif 'M' in x:
        x = x.replace('M', '')
        x = float(x) * 1000000
        return round(x)
    elif 'K' in x:
        x = x.replace('K', '')
        x = float(x) * 1000
        return round(x)
    else:
        return int(x)
    
data['Value'] = data['Value'].apply(convert_Col_Value_RC)

data
```

```python
data['Release Clause'] = data['Release Clause'].apply(convert_Col_Value_RC)
data
```

```python
data['Wage'].unique()
```

```python
#Convert Wage
def convert_wage(x):
    if "€" in x:
        return x[1:]
    elif "K" in x:
        x = x.replace("K", "")
        x = float(x) * 1000
        return round(x)
    else:
        return int(x)

data['Wage'] = data['Wage'].apply(convert_wage)
data
    
```

```python
data['Hits'].unique()
```

```python
#Converting Hits

def convert_hits(x):
    if "K" in x:
        x = x.replace("K", "")
        x = float(x) * 1000
        return round(x)
    elif x:
        return int(x)
    else:
        return 0

data['Hits'] = data['Hits'].apply(convert_hits)
```
