```python
import pandas as pd
import numpy as np
suicides_data = pd.read_csv("/Users/greshashah126/Desktop/master.csv", low_memory=False)
pd.set_option("display.max_columns", None)
suicides_data
```

```python
suicides_data.isnull().sum()
```

```python
suicides_data['suicides_no'] = suicides_data['suicides_no'].fillna(0) 
suicides_data['HDI for year'] = suicides_data['HDI for year'].fillna(0) 
```

```python
suicides_data = suicides_data.drop(columns='country-year')
```

```python
# Convert to integers
suicides_data['suicides_no'] = suicides_data['suicides_no'].astype(int)
```

```python
suicides_data['suicides_no'].dtype
```

```python
suicides_data.columns
```

```python
suicides_data = suicides_data.rename(columns={' gdp_for_year ($) ': 'gdp_for_year($)'})
```

```python
suicides_data['gdp_for_year($)'].unique()
```

```python
# Function to remove 'E+' part from numbers
def remove_num(x):
    if "E+" in x:
        return 0
    else:
        return x

suicides_data['gdp_for_year($)'] =  suicides_data['gdp_for_year($)'].apply(remove_num)
suicides_data['gdp_for_year($)'] = suicides_data['gdp_for_year($)'].str.replace(',', '')
suicides_data
    
```

```python
suicides_data.isnull().sum()
```

```python
suicides_data['gdp_for_year($)'] = suicides_data['gdp_for_year($)'].fillna(0) 
suicides_data
```

```python
suicides_data['gdp_for_year($)'].dtype
```

```python
suicides_data['gdp_for_year($)'] = suicides_data['gdp_for_year($)'].astype(int)
```

```python
suicides_data['gdp_per_capita ($)'].dtype
```

```python
#Reducing decimal place 
suicides_data['gdp_per_capita ($)'] = suicides_data['gdp_per_capita ($)'].apply (lambda x: round(x, 2))
suicides_data['suicides/100k pop'] = suicides_data['suicides/100k pop'].apply (lambda x: round(x, 3))
suicides_data['HDI for year'] = suicides_data['HDI for year'].apply (lambda x: round(x, 3))
suicides_data
```

```python
suicides_data.to_csv("SuicidesData_cleaned.csv", index=False)
```
